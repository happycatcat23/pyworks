```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression as LR
```

```python
train = pd.read_csv("/Users/happycatcat523/pyworks/kaggle資料/titanic/train.csv")
test = pd.read_csv("/Users/happycatcat523/pyworks/kaggle資料/titanic/test.csv")
train.head()
```

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
train['Embarked'] = train['Embarked'].fillna('S')
test['Age'] = test['Age'].fillna(test['Age'].median())
test['Fare'] = test['Fare'].fillna(test['Fare'].median())
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
dummy_train = pd.get_dummies(train[['Sex', 'Embarked']])
dummy_test = pd.get_dummies(test[['Sex', 'Embarked']])
```

```python
train_two = pd.concat([train.drop(["Sex","Embarked"],axis = 1),dummy_train], axis = 1)
test_two = pd.concat([test.drop(["Sex", "Embarked"], axis = 1),dummy_test], axis = 1)
```

```python
train_three = train_two.drop(['PassengerId', 'Name', 'SibSp', 'Ticket', 'Cabin', 'Parch'], axis = 1)
test_three = test_two.drop(['PassengerId', 'Name', 'SibSp', 'Ticket', 'Cabin', 'Parch'], axis = 1)
```

```python
X_train = train_three.drop(['Survived'], axis = 1).values
y_train = train_three.Survived.values
```

```python
lsr = LR()
train_fit = lsr.fit(X_train, y_train)
```

```python
test_predict = train_fit.predict(test_three)
test_predict_proba = train_fit.predict_proba(test_three)
test_predict
```

```python
test_predict_proba
```

```python
PassengerId = np.array(test['PassengerId']).astype(int)
my_solution = pd.DataFrame(test_predict, PassengerId, columns = ['Survived'])
my_solution.to_csv("logistic05.csv", index_label = ['PassengerId'])
```
