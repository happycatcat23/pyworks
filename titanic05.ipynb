```python
#パッケージをインポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LogisticRegression as LR
```

```python
train = pd.read_csv("/Users/happycatcat523/pyworks/kaggle資料/titanic/train.csv")
test = pd.read_csv("/Users/happycatcat523/pyworks/kaggle資料/titanic/test.csv")
```

```python
train.info()
```

```python
train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna('S')
```

```python
dummy_train = pd.get_dummies(train[['Sex', 'Embarked']])
dummy_train.head()
```

```python
train_two = pd.concat([train.drop(['Sex', 'Embarked'], axis = 1),dummy_train], axis = 1)
train_two.head()
```

```python
train_two.isnull().sum()
```

```python
train_three = train_two.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis = 1)
```

```python
train_three.info()
```

```python
X_train = train_three.drop('Survived', axis = 1)
y_train = train_three.Survived

slr = LR()
slr.fit(X_train, y_train)
```

```python
test.isnull().sum()
```

```python
test["Age"] = test["Age"].fillna(test["Age"].median())
test["Fare"] = test["Fare"].fillna(test["Fare"].mean())
```

```python
dummy_test = pd.get_dummies(test[['Sex','Embarked']])
dummy_test.head()
```

```python
test_two = pd.concat([test.drop(['Sex', 'Embarked'], axis = 1),dummy_test], axis = 1)
test_two.head()
```

```python
bench_mark = train_three.Survived.sum() / train_three.Survived.count()

print("bench_mark : ", bench_mark)
print("training score : ", slr.score(X_train, y_train))
```

```python
test_three = test_two.drop(["PassengerId", "Ticket", "Cabin", "Name"] , axis = 1)
test_three.info()
```

```python
test_predict = slr.predict(test_three)
test_predict[:5]
```

```python
PassengerId = np.array(test["PassengerId"]).astype(int)
my_solution = pd.DataFrame(test_predict, PassengerId, columns = ["Survived"])
my_solution.to_csv('logistic_three.csv', index_label = ["PassengerId"])
```
